````python
import pandas as pd
import seaborn as sns
import random
lst = ['robot'] * random.randint(5, 15)
lst += ['human'] * random.randint(5, 15)
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
````

````python
data.loc[data['whoAmI'] == 'robot', 'robot'] = 1
data.loc[data['whoAmI'] != 'robot', 'robot'] = 0
data.loc[data['whoAmI'] == 'human', 'human'] = 1
data.loc[data['whoAmI'] != 'human', 'human'] = 0
data[['robot', 'human']]
````

````python
sns.histplot(data=data, x='robot', hue='human')
````
